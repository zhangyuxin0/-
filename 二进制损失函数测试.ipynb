```python
import torch 
import torch.nn.functional as F
a= torch.rand(6,512,512)
a= F.sigmoid(a)
print(a.shape)
target = torch.rand(6,512,512)
print(target.shape)
```

```python
import torch.nn as nn
loss2=nn.BCELoss()
loss2=loss2(a,target)
loss1=nn.BCELoss(weight=None, reduction='none')   # 加这些参数就不求平均了
loss1=loss1(a,target)
```

```python
print(loss2)
print(loss1)
print(loss1.mean())
```
